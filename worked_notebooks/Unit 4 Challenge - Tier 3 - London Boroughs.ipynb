**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
#Write function based on above instruction. Keep in mind data types.
def create_price_ratio(x):
    y1998 = float(x['Average_Price'][x['Year'] == 1998])
    y2018 = float(x['Average_Price'][x['Year'] == 2018])
    ratio = [y2018/y1998]
    return ratio

#Call function with Westminster Borough as filter
create_price_ratio(dfg[dfg['London_Borough'] == 'Westminster'])
```

```python
#Loop through all values in London_Borough column.
all_boroughs = {}

for i in set(dfg['London_Borough']):
    london_borough = dfg[dfg['London_Borough'] == i]
    all_boroughs[i] = create_price_ratio(london_borough)
    
#See results of price ratio per London_Borough via all_borough dictionary
all_boroughs
```

```python
#Convert dictionary all_boroughs into a dataframe. It will provide better reporting capabilities to solve business case.
borough_ratios = pd.DataFrame(all_boroughs)
borough_ratios
```

```python
#Transpose dataframe and output it
borough_ratios_T = borough_ratios.transpose()
print(borough_ratios_T.sort_values(0))
```

```python
#reset index to rename column headers
borough_ratios_T = borough_ratios_T.reset_index()
borough_ratios_T
```

```python
#Rename column headers to better articulate dataframe
borough_ratios_T = borough_ratios_T.rename(columns = {'index': 'London_Borough', 0: '2018 Price Ratio'})
borough_ratios_T.sort_values(by = '2018 Price Ratio', ascending = False)
```

```python
#Plot dataframe via bar chart
plot2 = borough_ratios_T.plot(kind = 'bar', x = 'London_Borough', y = '2018 Price Ratio', title = 'London Borough 2018 Price Ratio Comparison' )

#Give Y-Axis a name, Price
plot2.set_ylabel('Price Ratio in GBP')
plot2.set_ylim([0, 6.5])
```

```python
top10_boroughs = borough_ratios_T.sort_values(by = '2018 Price Ratio', ascending = False).head(10)
print(top10_boroughs)

plot_top10_boroughs = top10_boroughs.plot(kind = 'bar', x = 'London_Borough', y = '2018 Price Ratio', title = '2018 Top 10 Highest London Borough Price Ratio Increases')
plot_top10_boroughs.set_ylim([0, 6.5])
```
